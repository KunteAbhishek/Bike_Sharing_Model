## Data Preparation

```python
# creating a list and putting all category columns in to it and converting them to category data type

categ = ['Season', 'Year', 'Month', 'holiday', 'weekday', 'workingday', 'Weather']

for i in categ:
    bike[i] = bike[i].astype('category')
```

```python
bike.info()
```

```python
# For Linear model we need to create the dummy variable

bike_dummy = pd.get_dummies(bike[categ], drop_first=True)
bike_dummy.head()
```

```python
# Now combining both dataframes bike and bike_dummy

bike_1 = pd.concat([bike, bike_dummy], axis=1)
bike_1.head()
```

```python
# Now we will drop columns from which we created dummy variables

bike_1.drop(categ, axis=1, inplace=True)
bike_1.head()
```

```python
bike_1.info()
```

### Now We will start model Building 

## Splitting the Data into Training and Testing Sets
As you know, the first basic step for regression is performing a train-test split.

```python
# Importing important libraries for model building

import sklearn
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
```

```python
# spliting the data into train and test dataframes

# We specify this so that the train and test data set always have the same rows, respectively
np.random.seed(0)

train_bike_1, test_bike_1 = train_test_split(bike_1, train_size = 0.70, random_state=100)
```

```python
# Checking shape of new dataframes

print(train_bike_1.shape)
print(test_bike_1.shape)
```

### Rescaling the Features 

 it is extremely important to rescale the variables so that they have a comparable scale. If we don't have comparable scales, then some of the coefficients as obtained by fitting the regression model might be very large or very small as compared to the other coefficients. This might become very annoying at the time of model evaluation. So it is advised to use standardization or normalization so that the units of the coefficients obtained are all on the same scale. As you know, there are two common ways of rescaling:

1. Min-Max scaling 
2. Standardisation (mean-0, sigma-1) 

This time, we will use MinMax scaling.

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# defining a variable scaler for minmax scaling

scaler = MinMaxScaler()
```

```python
bike_1.columns
```

```python
# Apply scaler() to all the columns except the 'yes-no' and 'dummy' variables

num_vars = ['temp', 'atemp', 'Humidity', 'windspeed', 'Count']

train_bike_1[num_vars] = scaler.fit_transform(train_bike_1[num_vars])
train_bike_1[num_vars].head()
```

```python
# checking all columns and all the variables after scaling

train_bike_1.describe()
```

```python
# Let's check the correlation coefficients to see which variables are highly correlated

plt.figure(figsize=(26,20))
sns.heatmap(train_bike_1.corr(), annot=True, cmap='YlGnBu')
plt.show()
```

### Insight: 
     The heatmap clearly shows which all variable are multicollinear in nature, and which variable have high collinearity with the target variable.

## Model Building

```python
# Dividing training set into X_train and y_train sets for the model building

y_train = train_bike_1.pop('Count')
X_train = train_bike_1

print(y_train.shape)
print(X_train.shape)
```

```python
# importing RFE library for feature selection and after this will perform manual feature selection.
# This time, we will be using the LinearRegression function from SciKit Learn for its compatibility with RFE (which is a utility from sklearn)

from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Create an instance of LinearRegression
lm = LinearRegression()

# Fit the linear regression model to the training data
lm.fit(X_train, y_train)

# Create an instance of RFE with desired number of features (15 in this case)
rfe = RFE(estimator=lm, n_features_to_select=15)

# Fit RFE to the training data
rfe.fit(X_train, y_train)
```

```python
#creating a list of features selected by RFE

list(zip(X_train, rfe.support_, rfe.ranking_))
```

```python
# creating a list of features selected by RFE

rfe_col = X_train.columns[rfe.support_]
rfe_col
```

```python
# feature which are chosen by RFE during feature selection( so un-supported columns)

X_train.columns[~rfe.support_]
```

```python
# Creating new train dataframe with RFE selected features

X_train_rfe = X_train [rfe_col]
X_train_rfe.head()
```

## Model 1

```python
# Adding a constant

X_train_lm1 = sm.add_constant(X_train_rfe)

# Creating first fitted model

lr1 = sm.OLS(y_train, X_train_lm1).fit()

lr1.summary()
```

```python
# Check for the VIF values of the feature variables. 

from statsmodels.stats.outliers_influence import variance_inflation_factor

# Create a dataframe that will contain the names of all the feature variables and their respective VIFs
vif = pd.DataFrame()
vif['Features'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe.values, i) for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

## Residaul Analysis

```python
from sklearn.metrics import r2_score

r2_test = r2_score(y_test, y_test_pred)
r2_train = r2_score(y_train, y_train_pred)

print('Test data r**2 :',round((r2_test*100),2))
print('Train data r**2 :',round((r2_train*100),2))
```

### Adjusted R**2 Value for TEST

```python
# n for test data ,n1 for train data is number of rows
n = X_test.shape[0]
n1 = X_train_rfe8.shape[0]

# Number of features (predictors, p for test data, p1 for train data) is the number of columns
p = X_test.shape[1]
p1 = X_train_rfe8.shape[1]


# We find the Adjusted R-squared using the formula

adjusted_r2_test = 1-(1-r2_test)*(n-1)/(n-p-1)
adjusted_r2_train = 1-(1-r2_train)*(n1-1)/(n1-p1-1)

print('Test data adjusted r^2 :',round((adjusted_r2_test*100),2))
print('Train data adjusted r^2 :',round((adjusted_r2_train*100),2))
```
